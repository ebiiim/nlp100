### 11. タブをスペースに置換

### 12. 1列目をcol1.txtに，2列目をcol2.txtに保存

各行の1列目だけを抜き出したものをcol1.txtに，2列目だけを抜き出したものをcol2.txtとしてファイルに保存せよ．確認にはcutコマンドを用いよ．

```python
def _cut(file, field, delimiter):
    with open(file, 'r') as f:
        data =[l.replace(delimiter, ' ').split() for l in f.readlines()]
        col = '\n'.join([l[field] for l in data])
    return col

col1 = _cut('data/hightemp.txt', 0, '\t')
col2 = _cut('data/hightemp.txt', 1, '\t')
print(f'{col1}\n-----\n{col2}')

with open('data/col1.txt', 'w') as f:
    f.write(col1)

with open('data/col2.txt', 'w') as f:
    f.write(col2)
```

```python
%%sh
sed 's/\t/ /g' data/hightemp.txt | cut -f 1 -d ' '
echo '-----'
sed 's/\t/ /g' data/hightemp.txt | cut -f 2 -d ' '
```

### 13. col1.txtとcol2.txtをマージ

12で作ったcol1.txtとcol2.txtを結合し，元のファイルの1列目と2列目をタブ区切りで並べたテキストファイルを作成せよ．確認にはpasteコマンドを用いよ．

```python
def _paste(A, B):
    print('\n'.join([f'{a}\t{b}'.replace('\n','') for a, b in zip(A, B)]))

with open('data/col1.txt', 'r') as f:
    col1 = f.readlines()

with open('data/col2.txt', 'r') as f:
    col2 = f.readlines()

_paste(col1, col2)
```

```python
%%sh
paste data/col1.txt data/col2.txt
```

### 14. 先頭からN行を出力

自然数Nをコマンドライン引数などの手段で受け取り，入力のうち先頭のN行だけを表示せよ．確認にはheadコマンドを用いよ．


```python
def _head(file, n):
    with open(file, 'r') as f:
        return ''.join(f.readlines()[:n])

print(_head('data/hightemp.txt', 5))
```

```python
%%sh
N=5
head -n$N data/hightemp.txt
```

### 15. 末尾のN行を出力

自然数Nをコマンドライン引数などの手段で受け取り，入力のうち末尾のN行だけを表示せよ．確認にはtailコマンドを用いよ．


```python
def _tail(file, n):
    with open(file, 'r') as f:
        return ''.join(f.readlines()[-n:])

print(_tail('data/hightemp.txt', 5))
```

```python
%%sh
N=5
tail -n$N data/hightemp.txt
```

### 16. ファイルをN分割する

自然数Nをコマンドライン引数などの手段で受け取り，入力のファイルを行単位でN分割せよ．同様の処理をsplitコマンドで実現せよ．


```python
def _split(file, n):
    with open(file, 'r') as f:
        lines = list(filter(lambda x:x!='', f.read().split('\n')))
        div = int(len(lines)/n)
        result = [lines[idx:idx+n] for idx in range(0, len(lines), n)]
    return result

for each in _split('data/hightemp.txt', 5):
    for line in each:
        print(line)
    print('---')
```

```python
%%sh
split -l 5 data/hightemp.txt data/hightemp-
cat data/hightemp-aa
echo '---'
cat data/hightemp-ab
echo '---'
cat data/hightemp-ac
echo '---'
cat data/hightemp-ad
echo '---'
cat data/hightemp-ae
echo '---'
```

### 18. 各行を3コラム目の数値の降順にソート

各行を3コラム目の数値の逆順で整列せよ（注意: 各行の内容は変更せずに並び替えよ）．確認にはsortコマンドを用いよ（この問題はコマンドで実行した時の結果と合わなくてもよい）．


```python
def _sort(file: str, key: int, delimiter: str):
    with open(file, 'r') as f:
        data = [l.strip('\n').split(delimiter) for l in f.readlines()]
        data.sort(key=lambda x:x[key])
        data.reverse()
        ret = ''
        for d in data:
            ret += delimiter.join(d) + '\n'
    return ret

print(_sort('data/hightemp.txt', 2, '\t'))
```

```python
%%sh
sort -r -t$'\t' -k3,3 data/hightemp.txt
```

### 19. 各行の1コラム目の文字列の出現頻度を求め，出現頻度の高い順に並べる

各行の1列目の文字列の出現頻度を求め，その高い順に並べて表示せよ．確認にはcut, uniq, sortコマンドを用いよ．


```python
import json
with open('data/col1.txt', 'r') as f:
    data = [l.strip('\n') for l in f.readlines()]
    data_freq = [data.count(d) for d in data]
    dfs = [[d, f] for d, f in zip(data, data_freq)]
    uniq_dfs = []
    for df in dfs:
        if df not in uniq_dfs:
            uniq_dfs.append(df)
    uniq_dfs.sort(key=lambda x:x[1])
    uniq_dfs.reverse()
    # print(uniq_dfs)
    for df in uniq_dfs:
        print(df[0])  # shellの場合と順序が異なる
```

```python
%%sh
sed 's/\t/ /g' data/hightemp.txt | cut -f 1 -d ' ' | sort | uniq -c | sort -r | cut -f 5 -d ' '
```
